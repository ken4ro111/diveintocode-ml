```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data[: , : ]
y = iris.target[:]
```

```python
X = pd.DataFrame(data = X , columns = ["sepal_length" , "sepal_width" , "petal_length" , "petal_width"])
y = pd.DataFrame(data = y , columns = ["Species"])
df = pd.concat([X , y] , axis = 1)
df.head()
```

```python
X = df_cf.iloc[: , [0,1]].values
y = df_cf.iloc[: , 2].values
```

## 【問題3】前処理・訓練用データと検証用データの分割


```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(
            X , y , test_size = 0.25)
```

## 【問題4】前処理・標準化


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

## 【問題5】学習と推定


```python
from sklearn.neighbors import KNeighborsClassifier
knn_1 = KNeighborsClassifier(n_neighbors=1)
knn_3 = KNeighborsClassifier(n_neighbors=3)
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_1.fit(X_train , y_train)
knn_3.fit(X_train , y_train)
knn_5.fit(X_train , y_train)
y_pred_1 = knn_1.predict(X_test)
y_pred_3 = knn_3.predict(X_test)
y_pred_5 = knn_5.predict(X_test)
```

```python
print("1-nnの推定結果{}".format(y_pred_1))
print("3-nnの推定結果{}".format(y_pred_3))
print("5-nnの推定結果{}".format(y_pred_5))
```

## 【問題6】評価


```python
#scikit-learnを使い4種類の指標を計算する
#Accuracy（正解率
#精度分類スコア。マルチラベル分類では、この関数はサブセットの精度を計算します。
#サンプルに対して予測されるラベルのセットは、y_trueの対応するラベルのセットと正確に一致する必要があります。

from sklearn.metrics import accuracy_score
acc_1 = accuracy_score(y_test , y_pred_1)
acc_3 = accuracy_score(y_test , y_pred_3)
acc_5 = accuracy_score(y_test , y_pred_5)
```

```python
#Precision（適合率）
#精度を計算する
#精度は比であり、真の陽性の数であり、偽陽性の数。
#精度とは、直感的には、分類器が負のサンプルを正とラベル付けしない能力です。
#tp / (tp + fp)tpfp   最適な値は1で、最悪の値は0です。

from sklearn.metrics import precision_score
pre_1 = precision_score(y_test , y_pred_1)
pre_3 = precision_score(y_test , y_pred_3)
pre_5 = precision_score(y_test , y_pred_5)
```

```python
#Recall（再現率
#リコールを計算する
#リコール比で真陽性の数とある偽陰性の数。
#リコールは直観的には、すべての陽性サンプルを見つける分類器の機能です.
#tp / (tp + fn)tpfn
#最適な値は1で、最悪の値は0です。

from sklearn.metrics import recall_score
rec_1 = recall_score(y_test , y_pred_1)
rec_3 = recall_score(y_test , y_pred_3)
rec_5 = recall_score(y_test , y_pred_5)
```

```python
#F値
#バランスFスコアまたはFメジャーとも呼ばれるF1スコアを計算
#F1スコアは、精度と再現率の加重平均として解釈できます。
#F1スコアは、1で最高値、0で最悪値に達します。
#F1スコアに対する精度と再現率の相対的な寄与は等しくなります。

from sklearn.metrics import f1_score
f_1 = f1_score(y_test , y_pred_1)
f_3 = f1_score(y_test , y_pred_3)
f_5 = f1_score(y_test , y_pred_5)
```

```python
#混同行列をscikit-learnを使い表示する


#混同行列（confusion matrix）
#混同行列を計算して分類の精度を評価する
#定義により、混同マトリックス  そのようなことです  
#グループ内にあることが知られている観測値の数に等しい  

from sklearn.metrics import confusion_matrix
mat_1 = confusion_matrix(y_test , y_pred_1)
mat_3 = confusion_matrix(y_test , y_pred_3)
mat_5 = confusion_matrix(y_test , y_pred_5)
```

```python
print("1-nnのAccuracy（正解率）は{}".format(acc_1))
print("1-nnのPrecision（適合率）は{}".format(pre_1))
print("1-nnのRecall（再現率）は{}".format(rec_1))
print("1-nnのF値は{}".format(f_1))
print("1-nnの混同行列（confusion matrix）は{}".format(mat_1))
```

# 4種類の指標値の意味について簡単に調査して文章でまとめる

真陽性 : TP (True-Positive)
ポジティブに分類すべきアイテムを正しくポジティブに分類できた件数を指します。
スパムフィルタの場合、「スパム」と分類すべきメールを「スパム」のフォルダに正しく分類できた件数を指します。

真陰性: TN (True-Negative)
ネガティブに分類すべきアイテムを正しくネガティブに分類できた件数を指します。
スパムフィルタの場合、「スパムでない」と分類すべきメールを「スパムでない」のフォルダに正しく分類できた件数を指します。

偽陽性: FP (False-Positive)
ネガティブに分類すべきアイテムを誤ってポジティブに分類した件数を指します。
スパムフィルタの場合、「スパムでない」と分類すべきメールを「スパム」のフォルダに誤って分類した件数を指します。

偽陰性: FN (False-Negative)
ポジティブに分類すべきアイテムを誤ってネガティブに分類した件数を指します。
スパムフィルタの場合、「スパム」に分類すべきメールを「スパムでない」のフォルダに誤って分類した件数を指します。

## 正解率（accuracy）

$$
accuracy = \frac{TP + TN} {TP + TN + FP + FN}
$$

本来陽性に分類すべきアイテムを陽性に分類し、本来陰性に分類すべきアイテムを陰性に分類できた割合

すべてのサンプルのうち正解したサンプルの割合。

## Precision（適合率）

$$
precision = \frac{TP}{TP + FP}
$$

精度 (Precision) とは、「陽性に分類されたアイテムのうち、実際に陽性であったアイテムの割合」

陽性と予測されたサンプルのうち正解したサンプルの割合

## Recall（再現率）

$$
recall = \frac{TP}{TP + FN}
$$

「本来陽性に分類すべきアイテムを、正しく陽性に分類できたアイテムの割合」を示し、以下の式で表されます。
検出率は、真陽性率 (TPR, True-Positive Rate) または、感度 (Sensitivity) とも呼ばれます。

## F値

$$
F-measure = \frac{2 * precision * recall}{precision + recall}
$$

精度 (Precision) と検出率 (Recall) をバランス良く持ち合わせているかを示す指標
精度は高くても検出率が低いモデルでないか、逆に検出率は高くても精度が低くなっていないかといった評価

## 【問題7】可視化


```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches

def decision_region(X, y, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は訓練データまたは検証データである。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        特徴量
    y : ndarray, shape(n_samples,)
        ラベル
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue']
    contourf_color = ['pink', 'skyblue']
    n_class = 2

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```
